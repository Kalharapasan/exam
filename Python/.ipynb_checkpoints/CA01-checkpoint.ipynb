```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("covid_datas.csv")

# Task 1: Display first five records and column names
print("First 5 Records:")
print(df.head())
print("\nColumn Names:")
print(df.columns.tolist())

# Task 2: Most frequently used “Illness_Severity”
print("\nIllness_Severity Counts:")
print(df["Illness_Severity"].value_counts())

# Task 3: Average of City_Code_Patient
avg_city_code = df["City_Code_Patient"].mean()
print(f"\nAverage City_Code_Patient: {avg_city_code:.2f}")

# Task 4: Distribution of patients across departments
print("\nDistribution Across Departments:")
print(df["Department"].value_counts())

# Task 5: Most common age groups
print("\nMost Common Age Groups:")
print(df["Age"].value_counts())

# Task 6: Number of unique patients
print(f"\nNumber of Unique Patients: {df['Patient_ID'].nunique()}")

# Task 7: Average Admission_Deposit per Type of Admission
print("\nAverage Admission_Deposit per Admission Type:")
print(df.groupby("Type_of_Admission")["Admission_Deposit"].mean())

# Task 8: Ward_Type with highest average Admission_Deposit
ward_avg = df.groupby("Ward_Type")["Admission_Deposit"].mean()
highest_ward = ward_avg.idxmax()
print(f"\nWard_Type with Highest Average Admission_Deposit: {highest_ward} ({ward_avg[highest_ward]:.2f})")

# Task 9.I: Bar chart - patient distribution across departments
plt.figure(figsize=(8, 5))
sns.countplot(x="Department", data=df)
plt.title("Patient Distribution Across Departments")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Task 9.II: Boxplot - Admission_Deposit by Department
plt.figure(figsize=(8, 5))
sns.boxplot(x="Department", y="Admission_Deposit", data=df)
plt.title("Admission Deposit Distribution by Department")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Task 9.III: Histogram - Distribution of Stay_Day
plt.figure(figsize=(7, 4))
sns.histplot(df["Stay_Day"], bins=20, kde=True)
plt.title("Distribution of Stay Day")
plt.xlabel("Stay Day")
plt.tight_layout()
plt.show()
```
