```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
df = sns.load_dataset('tips')  # built-in dataset with restaurant tips

# Set Seaborn style
sns.set(style="whitegrid")
```

```python
sns.histplot(data=df, x='total_bill')
plt.title('Histogram of Total Bill')
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=df)
plt.title('Average Total Bill by Day')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=df)
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
plt.show()
```

```python
sns.regplot(x='total_bill', y='tip', data=df)
plt.title('Regression Line for Tip vs Total Bill')
plt.show()
```

```python
sns.pairplot(df[['total_bill', 'tip', 'size']])
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Boxplot of Total Bill by Day')
plt.show()
```

```python
sns.swarmplot(x='day', y='total_bill', data=df)
plt.title('Swarm Plot of Total Bill by Day')
plt.show()
```

```python
sns.catplot(x='day', y='total_bill', kind='bar', data=df)
plt.title('Factor Plot (Catplot) - Total Bill by Day')
plt.show()
```
