```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [25, 30]
})

print(df.dtypes)     # Data types of each column
print(df.columns)    # Index(['Name', 'Age'])
print(df.axes)       # [RangeIndex(start=0, stop=2, step=1), Index(['Name', 'Age'], dtype='object')]
print(df.ndim)       # 2
print(df.size)       # 4 (2 rows Ã— 2 columns)
print(df.shape)      # (2, 2)
print(df.values)     # NumPy array representation
df.loc[0:3, ['Name', 'Age']]   # Rows 0-3, columns Name and Age
```

```python
df['Name']          # Single column as Series
df[['Name', 'Age']]       # Multiple columns as DataFrame
```
