```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Read MathStu.csv data using pandas
df = pd.read_csv("Math-Students.csv")

# 2. Find the Data types of every attribute
print("Data types:\n", df.dtypes)

# 3. Describe the statistics for each numeric column
print("Statistics:\n", df.describe())

# 4. Find the Standard Deviation of age
print("Standard Deviation of age:", df['age'].std())

# 5. Select first 10 data's from “famsize” column in the dataset
print("First 10 famsize values:\n", df['famsize'].head(10))

# 6. Find mean, median and Std Div for “freetime”
print("freetime - Mean:", df['freetime'].mean())
print("freetime - Median:", df['freetime'].median())
print("freetime - Std Dev:", df['freetime'].std())

# 7. Calculate mean age for each “address”
print("Mean age by address:\n", df.groupby('address')['age'].mean())

# 8. Select only those rows whose Mother's job (Mjob) is teacher
mjob_teacher = df[df['Mjob'] == 'teacher']
print("Rows where Mjob is teacher:\n", mjob_teacher)

# 9. Select the row in which age value is greater than 15
age_above_15 = df[df['age'] > 15]
print("Age > 15:\n", age_above_15)

# 10. Create new data from original sorted by column age
sorted_df = df.sort_values(by='age')
print("Sorted by age:\n", sorted_df)

# 11. Find Min, Max and Mean value of age and freetime
print("Age - Min:", df['age'].min(), "Max:", df['age'].max(), "Mean:", df['age'].mean())
print("Freetime - Min:", df['freetime'].min(), "Max:", df['freetime'].max(), "Mean:", df['freetime'].mean())

# 12. Draw Histogram for age data
sns.histplot(df['age'], bins=10)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# 13. Display Bar plot for Mjob and age
sns.barplot(x='Mjob', y='age', data=df)
plt.title("Bar Plot of Mjob vs Age")
plt.xticks(rotation=45)
plt.show()

# 14. Create Box plot for famsize and freetime
sns.boxplot(data=df[['famsize', 'freetime']])
plt.title("Box Plot for Famsize and Freetime")
plt.show()
```
